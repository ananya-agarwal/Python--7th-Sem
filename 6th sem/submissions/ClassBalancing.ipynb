```python
import pandas as pd
import numpy as np

test1 = pd.read_csv('/content/drive/MyDrive/Kaggle/h1-test_data.csv')
train1 = pd.read_csv('/content/drive/MyDrive/Kaggle/h1-train_data.csv')

train= train1.iloc[:,1:52]
test= test1.iloc[:,1:51]

```

```python
x = train.drop(['target'],axis=1)
y = train['target']
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=1000)
x_, y_ = smote.fit_resample(x, y)

print(x_[y_==0].shape)
print(x_[y_==1].shape)
```

```python
finaltrain = pd.concat([x_, y_],axis=1)
```

```python
evaluate_model(best)
```
