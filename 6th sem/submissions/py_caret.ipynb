```python
import pandas as pd
import numpy as np

test1 = pd.read_csv('/content/drive/MyDrive/Kaggle/h1-test_data.csv')
train1 = pd.read_csv('/content/drive/MyDrive/Kaggle/h1-train_data.csv')

train= train1.iloc[:,1:52]

test= test1.iloc[:,1:51]

train1.head()
```

```python
s = setup(train,target='target',remove_outliers = True)
```

```python
best = compare_models()
```

```python
evaluate_model(best)
```

```python
y_pred = predict_model(best,data=test)
```

```python
save_model(best,'my_best_pipeline')
```

```python
y_test0 = test1.iloc[:,0:1]
y_pred_label = y_pred.iloc[:,50:51]

y_pred_label.head()

y_test2 = pd.DataFrame(y_pred_label)
finallda = pd.concat([y_test0,y_test2], axis=1)
```
