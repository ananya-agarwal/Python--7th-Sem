```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("../input/2022-ucs654-Lab-2-Exam/h1-train_data.csv")
```

```python
X_train=df.iloc[:,1:51]
print(X_train)
```

```python
Y_train=df.iloc[:,51:52]
print(Y_train)
```

```python
# import support vector classifier 
# "Support Vector Classifier"

from sklearn.svm import SVC  
sv = SVC(kernel='linear') 
```

```python
df1 = pd.read_csv("../input/2022-ucs654-Lab-2-Exam/h1-test_data.csv")
```

```python
X_test=df1.iloc[:,1:51]
print(X_test)
```

```python
sv.fit(X_train, Y_train)
```

```python
y_test0 = df1.iloc[:,0:1]
y_test1 = sv.predict(X_test)
```

```python
y_test2 = pd.DataFrame(y_test1,columns = ['target'])
```

```python
final_df = pd.concat([y_test0,y_test2], axis=1)
```

```python
final_df.to_csv('svm.csv',index=False)
```
